```python
import pandas
dfs = pandas.read_html('https://rate.bot.com.tw/xrt?Lang=zh-TW')
```

```python
type(dfs)
```

```python
len(dfs)
```

```python
currency = dfs[0]
```

```python
currency = currency.ix[:,0:5]
```

```python
currency.columns = [u'幣別',u'現金匯率_本行買入',u'現金匯率_本行賣出',u'即期匯率_本行買入',u'即期匯率_本行賣出']
```

```python
currency
```

```python
currency[u'幣別']=currency[u'幣別'].str.extract('\((\w+)\)')
```

```python
currency
```

```python
currency.to_excel('台灣銀行牌告匯率.xlsx')
```
